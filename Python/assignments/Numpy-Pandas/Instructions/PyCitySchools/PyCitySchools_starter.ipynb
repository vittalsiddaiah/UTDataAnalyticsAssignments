```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
totalStudents = len(student_data['Student ID'].unique())
PctgMathPass = len(student_data[student_data.math_score >= 70])/totalStudents*100
PctgReadPass = len(student_data[student_data.reading_score >= 70])/totalStudents*100         
df = pd.DataFrame([
len(school_data['School ID'].unique()),
len(student_data['Student ID'].unique()),
school_data.budget.sum(),
student_data.math_score.mean(),
student_data.reading_score.mean(),
PctgMathPass,
PctgReadPass,
((PctgMathPass+PctgReadPass)/2.0)]).T
df.columns = ["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing Rate"]
df
```

## Top Performing Schools (By Passing Rate)

```python
df_school = school_data_complete.groupby(['school_name','type'])

df_totStudents = df_school.count().reset_index()
df_totStudents = df_totStudents.rename(columns={'Student ID': 'Total Students'})

df_totSchBudget = df_school.sum().reset_index()
df_totSchBudget = df_totSchBudget.rename(columns={'budget': 'Total School Budget'})

df_totStuBudget = df_school.mean().reset_index()
df_totStuBudget = df_totStuBudget.rename(columns={'budget': 'Per Student Budget'})
df_totStuBudget = df_totStuBudget.rename(columns={'math_score': 'Average Math Score'})
df_totStuBudget = df_totStuBudget.rename(columns={'reading_score': 'Average Reading Score'})
#school_data_complete.columns
#df_mathPassed = school_data_complete.columns
#df_readPassed = df_school[df_school.reading_score >=70]

df = df_totStudents.merge(df_totSchBudget,on='school_name', how='inner').merge(df_totStuBudget,on='school_name', how='inner')
df.sort_values(['Total School Budget'], ascending=False)
df.head()[['school_name', 'type',
           'Total Students', 
           'Total School Budget', 
           'Per Student Budget', 
           'Average Math Score',
           'Average Reading Score']]
```
